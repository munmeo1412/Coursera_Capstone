```python
 
# set width of bar
barWidth = 0.8
 
# set height of bar
bars1 = df['Very interested']
bars2 = df['Somewhat interested']
bars3 = df['Not interested']
 
# Set position of bar on X axis
r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
 
# Make the plot
plt.bar(r1, bars1, color='#5cb85c', width=barWidth, edgecolor='white', label='Very interested')
plt.bar(r2, bars2, color='#5bc0de', width=barWidth, edgecolor='white', label='Somewhat interested')
plt.bar(r3, bars3, color='#d9534f', width=barWidth, edgecolor='white', label='Not interested')
 
# Add xticks on the middle of the group bars
plt.xticks([r + barWidth for r in range(len(bars1))], ['Data Analysis / Statistics', 'Machine Learning',
       'Data Visualization', 'Big Data (Spark / Hadoop)', 'Deep Learning',
       'Data Journalism'])
 
# Create legend & Show graphic
plt.legend()
plt.title("Percentage of Respondents' Interest in Data Science Areas", fontsize=16)
plt.show()
```

```python
data = pd.read_csv('https://cocl.us/sanfran_crime_dataset')
print('Dataset downloaded and read into a pandas dataframe!')
```

```python
data.head()
```

```python
df_n=data.groupby(['PdDistrict'],as_index=False).count()
```

```python
df_n
```

```python
df_n.columns.values
```

```python
df_n=df_n.drop(['Category', 'Descript', 'DayOfWeek','Date', 'Time', 'Resolution', 'Address', 'X', 'Y', 'Location','PdId'],axis=1)
```

```python
df_n
```

```python
df_n.rename(columns={'PdDistrict':'Neighborhood','IncidntNum':'Count'}, inplace=True)
```

```python
df_n
```

```python
world_geo =gpd.read_file('https://cocl.us/sanfran_geojson') # geojson file

# create a plain world map
world_map = folium.Map(location=[37.773972, -122.431297], zoom_start=12)
```

```python
world_geo
```

```python
# generate choropleth map 
world_map.choropleth(
    geo_data=world_geo,
    data=df_n,
    columns=['Neighborhood', 'Count'],
    key_on='feature.properties.DISTRICT',
    fill_color='YlOrRd', 
    fill_opacity=0.7, 
    line_opacity=0.2,
    legend_name='Crime Rate in San Francisco'
)

# display map
world_map
```

```python


# create a numpy array of length 6 and has linear spacing from the minium total immigration to the maximum total immigration
threshold_scale = np.linspace(df_n['Count'].min(),
                              df_n['Count'].max(),
                              6, dtype=int)
threshold_scale = threshold_scale.tolist() # change the numpy array to a list
threshold_scale[-1] = threshold_scale[-1] + 1 # make sure that the last value of the list is greater than the maximum immigration

# let Folium determine the scale.
world_map = folium.Map(location=[37.773972, -122.431297], zoom_start=12)
world_map.choropleth(
    geo_data=world_geo,
    data=df_n,
    columns=['Neighborhood', 'Count'],
    key_on='feature.properties.DISTRICT',
    threshold_scale=threshold_scale,
    fill_color='YlOrRd', 
    fill_opacity=0.7, 
    line_opacity=0.2,
    legend_name='Crime Rate in San Francisco',
    reset=True
)
world_map
```
